```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
heart_df = pd.read_csv("/content/heart-disease.csv")
heart_df.head()
```

```python
heart_df.shape
```

```python
#Histogram of age
heart_df['age'].plot.hist(bins=50)
```

```python
heart_df.plot.hist(figsize = (10,30),subplots=True)
```

#Object Oriantated Method


```python
heart_df.dtypes
```

```python
over_50 = heart_df[heart_df['age']>50]
over_50.shape
```

```python
#OO method
fig,ax = plt.subplots(figsize=(10,6))

#Plot the data
scatter = ax.scatter(x=over_50['age'],
                     y=over_50['chol'],
                     c=over_50['target'])

#Customize the plot
ax.set(title="Heart disease and cholestrol level",
       xlabel = 'Age',
       ylabel = 'Cholestrol')

#Legend
ax.legend(*scatter.legend_elements(),title='Has heart disease')

#add horizontal line
ax.axhline(over_50['chol'].mean(),
           linestyle='--')
```

```python
fig,(ax0,ax1) = plt.subplots(figsize=(10,10),
                             nrows=2,
                             ncols=1,
                             sharex=True)

scatter = ax0.scatter(x = over_50['age'],
                      y = over_50['chol'],
                      c = over_50['target'])

ax0.set(title="Heart disease and cholestrol level",
       ylabel = 'Cholestrol')

ax0.legend(*scatter.legend_elements(),title='Has heart disease')

ax0.axhline(over_50['chol'].mean(),
           linestyle='--')

scatter = ax1.scatter(x = over_50['age'],
                      y = over_50['thalach'],
                      c = over_50['target'])

ax1.set(title="Heart disease and max heart rate",
        xlabel='Age',
       ylabel = 'thalach')

ax1.legend(*scatter.legend_elements(),title='Has heart disease')

ax1.axhline(over_50['thalach'].mean(),
           linestyle='--')

fig.suptitle("Heart Disease Analysis",fontsize=16,fontweight='bold')
```

```python
plt.style.use('seaborn-v0_8-darkgrid')
#OO method
fig,ax = plt.subplots(figsize=(10,6))

#Plot the data
scatter = ax.scatter(x=over_50['age'],
                     y=over_50['chol'],
                     c=over_50['target'],
                     cmap='winter')#change color scheme

#Customize the plot
ax.set(title="Heart disease and cholestrol level",
       xlabel = 'Age',
       ylabel = 'Cholestrol')

#Legend
ax.legend(*scatter.legend_elements(),title='Has heart disease')

#add horizontal line
ax.axhline(over_50['chol'].mean(),
           linestyle='--',
           c='red')
```

```python
#custmize x and y limitaions
fig,(ax0,ax1) = plt.subplots(figsize=(10,10),
                             nrows=2,
                             ncols=1,
                             sharex=True)

scatter = ax0.scatter(x = over_50['age'],
                      y = over_50['chol'],
                      c = over_50['target'],
                      cmap='summer')




ax0.set(title="Heart disease and cholestrol level",
       ylabel = 'Cholestrol')

ax0.set_xlim([50,80])
ax0.legend(*scatter.legend_elements(),title='Has heart disease')

ax0.axhline(over_50['chol'].mean(),
           linestyle='--',
            c='red')

scatter = ax1.scatter(x = over_50['age'],
                      y = over_50['thalach'],
                      c = over_50['target'],
                      cmap='winter')





ax1.set(title="Heart disease and max heart rate",
        xlabel='Age',
       ylabel = 'thalach')
ax1.set_xlim([50,80])

ax1.legend(*scatter.legend_elements(),title='Has heart disease')

ax1.axhline(over_50['thalach'].mean(),
           linestyle='--',
            c='red')

fig.suptitle("Heart Disease Analysis",fontsize=16,fontweight='bold')
```
